# Principal Component Analysis (PCA)
Principal Component Analysis (PCA) is a statistical technique used for dimensionality reduction, data compression, and feature extraction. It transforms the original dataset into a new coordinate system where the greatest variances lie on the first coordinates (called principal components), and each subsequent component is orthogonal to the previous ones.


```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Sample data (e.g., 3 features)
X = np.array([
    [2.5, 2.4, 1.0],
    [0.5, 0.7, 0.5],
    [2.2, 2.9, 1.5],
    [1.9, 2.2, 1.3],
    [3.1, 3.0, 2.2],
    [2.3, 2.7, 1.7],
    [2, 1.6, 1.0],
    [1, 1.1, 0.7],
    [1.5, 1.6, 1.1],
    [1.1, 0.9, 0.5]
])

# Standardize the data (optional)
X_mean = np.mean(X, axis=0)
X_centered = X - X_mean

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_centered)

print("Original Data:")
print(X)
print("\nTransformed Data (PCA):")
print(X_pca)

# Plot the transformed data
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Result')
plt.show()
```
