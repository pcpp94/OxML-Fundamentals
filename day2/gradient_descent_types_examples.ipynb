# Gradient Descent (GD)

```python
import numpy as np
import matplotlib.pyplot as plt

# Objective function
def f(x):
    return (x - 3) ** 2

# Gradient of the objective function
def f_prime(x):
    return 2 * (x - 3)

# Gradient Descent function
def gradient_descent(starting_x, learning_rate, epochs):
    x = starting_x
    history = []  # To store x values through the training process
    for _ in range(epochs):
        grad = f_prime(x)
        x = x - learning_rate * grad
        history.append(x)
    return x, history

# Parameters
starting_x = 0  # Starting point
learning_rate = 0.1
epochs = 25

# Perform gradient descent
final_x, history = gradient_descent(starting_x, learning_rate, epochs)

# Print the results
print(f"Optimal x value: {final_x}")
print(f"Function value at optimal x: {f(final_x)}")

# Plot the function and the gradient descent path
x_values = np.linspace(-1, 5, 400)
y_values = f(x_values)
plt.plot(x_values, y_values, label='f(x) = (x-3)^2')
plt.scatter(history, [f(x) for x in history], color='red', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Gradient Descent Optimization')
plt.show()
```



In the context of the diagrams showing the effect of different learning rates (\(\eta\)) on the gradient descent optimization process, $C$ typically represents the curvature of the cost function near the minimum, which is related to the second derivative (or Hessian in higher dimensions) of the cost function. More formally, it can be thought of as:

### Interpretation of $C$

1. **Curvature**:
   - $C$ represents the curvature of the objective function at the point of optimization.
   - In simple terms, it can be understood as a measure of how steep or flat the curve is around the minimum.

2. **Second Derivative**:
   - For a quadratic cost function $f(w) = \frac{1}{2} C w^2$, $C$ is the second derivative of the cost function with respect to the parameter $w$, i.e., $C = \frac{d^2 f(w)}{dw^2}$.
   - In higher dimensions, $C$ would be analogous to the largest eigenvalue of the Hessian matrix $\nabla^2 f(w)$.

### Effect on Learning Rate

- **Optimal Learning Rate**:
  $$
  \eta = \frac{1}{C}
  $$
  This value ensures the gradient descent step moves directly towards the minimum in an efficient manner without overshooting.

- **Small Learning Rate**:
  $$
  \eta < \frac{1}{C}
  $$
  This makes the steps too small, resulting in slow convergence.

- **Oscillatory Learning Rate**:
  $$
  \frac{1}{C} < \eta < \frac{2}{C}
  $$
  This range can cause the steps to oscillate around the minimum but still converge eventually.

- **Divergent Learning Rate**:
  $$
  \eta > \frac{2}{C}
  $$
  This results in steps that are too large, causing the algorithm to diverge.

### Summary

- $C$ is related to the curvature or the second derivative of the cost function.
- It helps in determining the optimal learning rate for gradient descent to ensure efficient and stable convergence.

# Stochastic Gradient Descent (SGD)

```python
import numpy as np
import matplotlib.pyplot as plt

# Objective function
def f(x):
    return (x - 3) ** 2

# Gradient of the objective function
def f_prime(x):
    return 2 * (x - 3)

# Stochastic Gradient Descent function
def stochastic_gradient_descent(starting_x, learning_rate, epochs, batch_size):
    x = starting_x
    history = []  # To store x values through the training process
    for _ in range(epochs):
        for _ in range(batch_size):
            grad = f_prime(x)
            x = x - learning_rate * grad
            history.append(x)
    return x, history

# Parameters
starting_x = 0  # Starting point
learning_rate = 0.1
epochs = 25
batch_size = 1  # For true SGD, use a batch size of 1

# Perform stochastic gradient descent
final_x, history = stochastic_gradient_descent(starting_x, learning_rate, epochs, batch_size)

# Print the results
print(f"Optimal x value: {final_x}")
print(f"Function value at optimal x: {f(final_x)}")

# Plot the function and the gradient descent path
x_values = np.linspace(-1, 5, 400)
y_values = f(x_values)
plt.plot(x_values, y_values, label='f(x) = (x-3)^2')
plt.scatter(history, [f(x) for x in history], color='red', label='SGD Path', s=10)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Stochastic Gradient Descent Optimization')
plt.show()
```

# Mini-SGD

```python
import numpy as np
import matplotlib.pyplot as plt

# Objective function
def f(x):
    return (x - 3) ** 2

# Gradient of the objective function
def f_prime(x):
    return 2 * (x - 3)

# Mini-Batch Stochastic Gradient Descent function
def mini_sgd(starting_x, learning_rate, epochs, batch_size):
    x = starting_x
    history = []  # To store x values through the training process
    data_points = np.linspace(-1, 5, 100)  # Simulating a dataset
    
    for _ in range(epochs):
        np.random.shuffle(data_points)  # Shuffle the dataset
        for i in range(0, len(data_points), batch_size):
            batch = data_points[i:i + batch_size]
            grad = np.mean([f_prime(xi) for xi in batch])
            x = x - learning_rate * grad
            history.append(x)
    
    return x, history

# Parameters
starting_x = 0  # Starting point
learning_rate = 0.1
epochs = 25
batch_size = 10  # Mini-batch size

# Perform mini-batch SGD
final_x, history = mini_sgd(starting_x, learning_rate, epochs, batch_size)

# Print the results
print(f"Optimal x value: {final_x}")
print(f"Function value at optimal x: {f(final_x)}")

# Plot the function and the gradient descent path
x_values = np.linspace(-1, 50, 400)
y_values = f(x_values)
plt.plot(x_values, y_values, label='f(x) = (x-3)^2')
plt.scatter(history, [f(x) for x in history], color='red', label='Mini-SGD Path', s=10)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Mini-Batch Stochastic Gradient Descent Optimization')
plt.show()
```

# Adam GD (adaptive moment estimation)

```python
import numpy as np
import matplotlib.pyplot as plt

# Objective function
def f(x):
    return (x - 3) ** 2

# Gradient of the objective function
def f_prime(x):
    return 2 * (x - 3)

# Adam optimizer function
def adam(starting_x, learning_rate, epochs, beta1, beta2, epsilon):
    x = starting_x
    history = []  # To store x values through the training process
    m = 0
    v = 0
    t = 0
    
    for _ in range(epochs):
        t += 1
        grad = f_prime(x)
        m = beta1 * m + (1 - beta1) * grad
        v = beta2 * v + (1 - beta2) * (grad ** 2)
        
        m_hat = m / (1 - beta1 ** t)
        v_hat = v / (1 - beta2 ** t)
        
        x = x - learning_rate * m_hat / (np.sqrt(v_hat) + epsilon)
        history.append(x)
    
    return x, history

# Parameters
starting_x = 0  # Starting point
learning_rate = 0.1
epochs = 100
beta1 = 0.9
beta2 = 0.999
epsilon = 1e-8

# Perform Adam optimization
final_x, history = adam(starting_x, learning_rate, epochs, beta1, beta2, epsilon)

# Print the results
print(f"Optimal x value: {final_x}")
print(f"Function value at optimal x: {f(final_x)}")

# Plot the function and the gradient descent path
x_values = np.linspace(-1, 5, 400)
y_values = f(x_values)
plt.plot(x_values, y_values, label='f(x) = (x-3)^2')
plt.scatter(history, [f(x) for x in history], color='red', label='Adam Path', s=10)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Adam Optimization')
plt.show()
```
